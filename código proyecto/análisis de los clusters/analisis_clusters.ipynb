### Automatizando 

```python
import pickle
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

def get_afinity_data (partition):
    data = None
    if partition == 'men': 
        data = men_df
    elif partition == 'women': 
        data = women_df
    elif partition == 'all':
        data = all_df
    return data

partitions = ['men', 'women', 'all']
n_cluster_sizes = [2, 3, 4, 5, 10, 15, 20, 30, 40, 50, 60, 70 ,80 ,90, 100]
results = {}

for partition in partitions:
    data = get_afinity_data(partition)
    remover = ['timestamp', 'item_id', 'position', 'click', 'propensity_score', 'user_feature_0', 'user_feature_1', 'user_feature_2', 'user_feature_3']
    user_item_affinities = data.drop(columns=remover)
    print(f"\n--------------------- Particion: {partition} ---------------------\n")
    for n_clusters in n_cluster_sizes:
        try:
            print(f"\n--------------------- KMeans con {n_clusters} clusters ---------------------\n")
            results[n_clusters] = {}
            kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(user_item_affinities)
            user_item_affinities['cluster'] = kmeans.labels_

            cluster_shapes = []
            for i in range(n_clusters):
                cluster = user_item_affinities[user_item_affinities['cluster'] == i]
                print("Tamaño cluster ", i, ": ", cluster.shape)
                cluster_shapes.append(cluster.shape)

            X = user_item_affinities.drop(columns=['cluster'])      
            y = user_item_affinities['cluster']

            results[n_clusters]['shapes'] = cluster_shapes
            results[n_clusters]['silhouette'] = silhouette_score(X, y, metric='euclidean')
            
            print(f"Silhouette score para {n_clusters} clusters: {results[n_clusters]['silhouette']}")
            
        except Exception as e:
            print(f"Iteracion detenida en {n_clusters} clusters, continuando  con la siguiente particion...\n")
            print(f"Error: {e}")
            break

    with open(f'results_{partition}_affinities.pkl', 'wb') as f:
        pickle.dump(results, f)
```

```python
# Los resultados tienen la forma: 

# { n_clusters_1: { 
#                 'shapes': [ (shape_cluster_0), (shape_cluster_1), ...],
#                 'silhouette': silhouette_score 
#              },  
#   n_clusters_2: { 
#                 'shapes': [ (shape_cluster_0), (shape_cluster_1), ...],
#                 'silhouette': silhouette_score
#              },
#   ...
# } 

# Ahora exploramos cada cluster para cada particion, identificando 
# el mejor silhouette score

def get_best_cluster(results):
    best_cluster = None
    best_score = -1
    # results.items() retorna una lista de tuplas (key, value)
    for n_clusters, data in results.items():
        silhouette = data['silhouette']
        if silhouette > best_score:
            best_score = silhouette
            best_cluster = n_clusters
    return best_cluster, best_score

# También obtenemos el tamaño minimo, maximo y promedio de 
# una clusterización particular
def get_cluster_size_stats(results, n_cluster):
    cluster_sizes = []
    shapes = results['n_cluster']['shapes']
    for shape in shapes:
        cluster_sizes.append(shape[0])
    return min(cluster_sizes), max(cluster_sizes), sum(cluster_sizes)/n_clusters
```

```python
import pickle
from pprint import pprint

def process_clusters_silhouette(category, affinity_results, features_results):

    processed_data = {category: {}}
    
    processed_data[category]['affinity'] = {}
    best_cluster, best_score = get_best_cluster(affinity_results)
    processed_data[category]['affinity']['best_cluster_size'] = best_cluster
    processed_data[category]['affinity']['best_cluster_silouette'] = best_score
    
    processed_data[category]['features'] = {}
    best_cluster, best_score = get_best_cluster(features_results)
    processed_data[category]['features']['best_cluster_size'] = best_cluster
    processed_data[category]['features']['best_cluster_silouette'] = best_score
    
    return processed_data


processed_clusters = {}

categories = ['men', 'women', 'all']
results = {
    'men': {'affinity': men_affinity_results, 'features': men_features_results},
    'women': {'affinity': women_affinity_results, 'features': women_features_results},
    'all': {'affinity': all_affinity_results, 'features': all_features_results},
}

for category in categories:
    cluster_data = process_clusters_silhouette(category, results[category]['affinity'], results[category]['features'])
    processed_clusters.update(cluster_data)

with open('processed_clusters.pkl', 'wb') as f:
    pickle.dump(processed_clusters, f)

pprint(processed_clusters)
```

```python
# Los resultados tienen la forma: 

# results: 
# { 
#   n_clusters_1: [df_cluster_1, df_cluster_2, ... df_cluster_n], 
#   n_clusters_2: [df_cluster_1, df_cluster_2, ... df_cluster_n],
#   ...
# }

# size_results:
# { 
#   n_clusters_1: [tamaño cluster 1, tamaño cluster 2, ... tamaño cluster n], 
#   n_clusters_2: [tamaño cluster 1, tamaño cluster 2, ... tamaño cluster n],
#   ...
# }

# Ahora exploramos cada cluster para cada particion, identificando
# el mejor estimador OPE

# Para acceder a los resultados de n_clusters = 2, cluster = 0, estimador = 'ipw': 
# results_men[2][0].loc['ipw'][0]

def get_weighted_average(estimators, sizes):
    weighted_average = 0
    for i in range(len(estimators)):
        weighted_average += estimators[i] * sizes[i]
    return weighted_average / sum(sizes)


def get_best_estimator(results, size_results):
    best_ipw = None
    best_dm = None
    best_dr = None

    best_ipw_score = -1
    best_dm_score = -1
    best_dr_score = -1

    for n_clusters, data in results.items():
        ipw_estimators = []
        dm_estimators = []
        dr_estimators = []

        for i in range(len(data)):
            ipw_estimators.append(data[i].loc['ipw'][0])
            dm_estimators.append(data[i].loc['dm'][0])
            dr_estimators.append(data[i].loc['dr'][0])


        ipw_score = get_weighted_average(ipw_estimators, size_results[n_clusters])
        dm_score = get_weighted_average(dm_estimators, size_results[n_clusters])
        dr_score = get_weighted_average(dr_estimators, size_results[n_clusters])

        if ipw_score > best_ipw_score:
            best_ipw_score = ipw_score
            best_ipw = n_clusters

        if dm_score > best_dm_score:
            best_dm_score = dm_score
            best_dm = n_clusters

        if dr_score > best_dr_score:
            best_dr_score = dr_score
            best_dr = n_clusters
    
    return {
        "ipw": (best_ipw, best_ipw_score),
        "dm": (best_dm, best_dm_score),
        "dr": (best_dr, best_dr_score)
        }
```

### Graficos n_clusters v/s silhouette & OPE

#### OPE

```python
# Primero obtenemos los promedios ponderados para cada n_cluster en los resultados

def get_averages(results, size_results):
    sizes=[]
    averages = []

    for n_clusters, data in results.items():

        ipw_estimators = []
        dm_estimators = []
        dr_estimators = []

        for i in range(len(data)):
            ipw_estimators.append(data[i].loc['ipw'][0])
            dm_estimators.append(data[i].loc['dm'][0])
            dr_estimators.append(data[i].loc['dr'][0])


        ipw_score = get_weighted_average(ipw_estimators, size_results[n_clusters])
        dm_score = get_weighted_average(dm_estimators, size_results[n_clusters])
        dr_score = get_weighted_average(dr_estimators, size_results[n_clusters])

        sizes.append(n_clusters)
        averages.append((ipw_score, dm_score, dr_score))
    
    return sizes, averages
```

```python
OPE_averages = {}

OPE_averages["men_affinity"] = get_averages(results_men_affinity, size_results_men_affinity)
OPE_averages["women_affinity"] = get_averages(results_women_affinity, size_results_women_affinity)
OPE_averages["all_affinity"] = get_averages(results_all_affinity, size_results_all_affinity)
OPE_averages["men_features"] = get_averages(results_men, size_results_men)
OPE_averages["women_features"] = get_averages(results_women, size_results_women)
OPE_averages["all_features"] = get_averages(results_all, size_results_all)

pprint(OPE_averages)
with open('OPE_averages.pkl', 'wb') as f:
    pickle.dump(OPE_averages, f)
```

```python
men_affinity_averages = get_averages(results_men, size_results_men)
import matplotlib.pyplot as plt

def plot_weighted_averages(data, title, baseline=None):
    sizes, averages = get_averages(*data)

    # Separate the averages into IPW, DM, and DR
    ipw_averages = [avg[0] for avg in averages]
    dm_averages = [avg[1] for avg in averages]
    dr_averages = [avg[2] for avg in averages]

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, ipw_averages, label='IPW', marker='o')
    plt.plot(sizes, dm_averages, label='DM', marker='o')
    plt.plot(sizes, dr_averages, label='DR', marker='o')

    if baseline is not None:
        plt.axhline(y=baseline, color='r', linestyle='--', label='Baseline')

    plt.xlabel('Number of Clusters')
    plt.ylabel('Weighted Average')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.savefig(f"Graficos/{title}.png")

plot_weighted_averages((results_men_affinity, size_results_men_affinity), 'OPE vs #Clusters: Men Affinity')
plot_weighted_averages((results_men, size_results_men), 'OPE vs #Clusters: Men features')

plot_weighted_averages((results_women_affinity, size_results_women_affinity), 'OPE vs #Clusters: Women Affinity')
plot_weighted_averages((results_women, size_results_women), 'OPE vs #Clusters: Women features')

plot_weighted_averages((results_all_affinity, size_results_all_affinity), 'OPE vs #Clusters: All Affinity')
plot_weighted_averages((results_all, size_results_all), 'OPE vs #Clusters: All features')


```

```python
def get_best_clusters(n_cluster, results, size_results, metric = None): 

    baseline = {
        "ipw": 0.006,
        "dm": 0.006,
        "dr": 0.006
    }

    cluster_cuantity = len(results[n_cluster])
    results = results[n_cluster]
    size_results = size_results[n_cluster]
    best_clusters = []

    for i in range(cluster_cuantity):
        size = size_results[i]
        ipw = results[i].loc['ipw'][0]
        dm = results[i].loc['dm'][0]
        dr = results[i].loc['dr'][0]
        # print(f"Cluster {i}:")
        # print(f"IPW: {ipw}")
        # print(f"DM: {dm}")
        # print(f"DR: {dr}")
        # print("\n")

        if metric == 'ipw':
            if ipw > baseline['ipw']:
                best_clusters.append((i, ipw))
        
        elif metric == 'dm':
            if dm > baseline['dm']:
                best_clusters.append((i, dm))
        
        elif metric == 'dr':
            if dr > baseline['dr']:
                best_clusters.append((i, dr))

        elif ipw > baseline['ipw'] or dm > baseline['dm'] or dr > baseline['dr']:
            best_clusters.append((i, size, ipw, dm, dr))

    return best_clusters
```

```python
results_list = [results_men, results_men_affinity, results_women, results_women_affinity, results_all, results_all_affinity]
size_list = [size_results_men, size_results_men_affinity, size_results_women, size_results_women_affinity, size_results_all, size_results_all_affinity]
results_labels = ['men_features', 'men_affinities', 'women_features', 'women_affinities', 'all_features', 'all_affinities']
best_clusters = {}
best_clusters_cuantities = {}

for i in range(len(results_list)):
    result = results_list[i]
    size_result = size_list[i]
    label = results_labels[i]

    best_clusters[label] = {}
    for n_clusters, data in result.items():
        best_clusters[label][n_clusters] = get_best_clusters(n_clusters, result, size_result)
        print(f"Best clusters for {label} with {n_clusters} clusters: {best_clusters[label][n_clusters]}")

for i in range(len(results_list)):
    result = results_list[i]
    size_result = size_list[i]
    label = results_labels[i]

    best_clusters_cuantities[label] = {}
    for n_clusters, data in result.items():
        best_clusters_cuantities[label][n_clusters] = len(get_best_clusters(n_clusters, result, size_result))
        print(f"Nº good clusters for {label} with {n_clusters} clusters: {best_clusters_cuantities[label][n_clusters]}")
```
